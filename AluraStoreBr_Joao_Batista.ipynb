#1. Análise do faturamento


```python
def analisar_faturamento(loja, nome_loja):

  faturamento = loja['Preço'].sum()
  print(f'Faturamento Total da {nome_loja}: {faturamento}')
  return faturamento

faturamento_loja1 = analisar_faturamento(loja, 'Loja 1')
faturamento_loja2 = analisar_faturamento(loja2, 'Loja 2')
faturamento_loja3 = analisar_faturamento(loja3, 'Loja 3')
faturamento_loja4 = analisar_faturamento(loja4, 'Loja 4')
```

```python
import matplotlib.pyplot as plt

def gerar_grafico_pizza(lojas_dados):
    nomes_lojas = []
    valores_faturamento = []

    for loja, nome_loja in lojas_dados:
        nomes_lojas.append(nome_loja)
        valores_faturamento.append(loja['Preço'].sum())

    # --- Personalização para mostrar VALORES (R$) ---
    def formato_absoluto(val):
        total = sum(valores_faturamento)
        valor_absoluto = int(val/100.*total)
        return f'R$ {valor_absoluto:,}'.replace(",", ".")  # Formato: R$ 1.000

    plt.figure(figsize=(10, 6))
    cores = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
    explode = (0.05, 0.05, 0.05, 0.05)

    plt.pie(
        valores_faturamento,
        labels=nomes_lojas,
        colors=cores,
        autopct=formato_absoluto,  # Aqui usamos a função personalizada!
        startangle=90,
        explode=explode,
        shadow=True,
        textprops={'fontsize': 10}
    )

    plt.title('Faturamento Total por Loja (Valores Absolutos)', fontsize=14)
    plt.tight_layout()
    plt.savefig('faturamento_pizza_lojas.png', dpi=300, bbox_inches='tight')
    plt.show()

# Uso:
lojas = [(loja, 'Loja 1'), (loja2, 'Loja 2'), (loja3, 'Loja 3'), (loja4, 'Loja 4')]
gerar_grafico_pizza(lojas)
```

# 2. Vendas por Categoria


```python
def faturamento_por_categoria_mais_populares(loja, nome_loja):

    quantidade_categoria = loja.groupby('Categoria do Produto').size().reset_index(name='Quantidade de Vendas')

    '''
groupby('Categoria do Produto'): Agrupa os dados pela coluna Categoria do Produto.
.size(): Conta quantos produtos foram vendidos em cada categoria.
.reset_index(name="Quantidade de Vendas"): Transforma o resultado em uma nova tabela com duas colunas:
Categoria do Produto (nome da categoria).
Quantidade de Vendas (total de produtos vendidos nela).
'''

    quantidade_categoria = quantidade_categoria.sort_values(by='Quantidade de Vendas', ascending=False)

    print(f'\nPopularidade de categorias em ordem decrescente - {nome_loja}:')
    print(quantidade_categoria)

    #print(type(quantidade_categoria))

    return quantidade_categoria



faturamento_loja1 = faturamento_por_categoria_mais_populares(loja, 'Loja 1')
faturamento_loja2 = faturamento_por_categoria_mais_populares(loja2, 'Loja 2')
faturamento_loja3 = faturamento_por_categoria_mais_populares(loja3, 'Loja 3')
faturamento_loja4 = faturamento_por_categoria_mais_populares(loja4, 'Loja 4')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def grafico_dispersao_vendas_faturamento(loja, nome_loja):
    """
    Gera um gráfico de dispersão entre quantidade de vendas e faturamento por categoria.

    Parâmetros:
    loja (DataFrame): Dados da loja.
    nome_loja (str): Nome da loja para exibição.
    """
    # Processamento dos dados
    dados_agrupados = loja.groupby('Categoria do Produto').agg(
        Quantidade=('Categoria do Produto', 'size'),
        Faturamento=('Preço', 'sum')
    ).reset_index()

    # Normaliza os valores para o tamanho dos pontos (5 a 100)
    tamanho_min, tamanho_max = 5, 100
    tamanhos = np.interp(dados_agrupados['Faturamento'],
                        (dados_agrupados['Faturamento'].min(),
                         dados_agrupados['Faturamento'].max()),
                        (tamanho_min, tamanho_max))

    # Criação do gráfico
    plt.figure(figsize=(12, 7))

    # Scatter plot com cores e tamanhos dinâmicos
    scatter = plt.scatter(
        x=dados_agrupados['Quantidade'],
        y=dados_agrupados['Faturamento'],
        s=tamanhos,  # Tamanho proporcional ao faturamento
        c=dados_agrupados['Faturamento'],  # Cor proporcional ao faturamento
        cmap='viridis',
        alpha=0.7,
        edgecolors='black',
        linewidth=0.5
    )

    # Adiciona rótulos aos pontos
    for i, row in dados_agrupados.iterrows():
        plt.annotate(
            row['Categoria do Produto'],
            (row['Quantidade'], row['Faturamento']),
            textcoords="offset points",
            xytext=(0, 10),
            ha='center',
            fontsize=9
        )

    # Linha de tendência
    z = np.polyfit(dados_agrupados['Quantidade'], dados_agrupados['Faturamento'], 1)
    p = np.poly1d(z)
    plt.plot(dados_agrupados['Quantidade'],
             p(dados_agrupados['Quantidade']),
             color='red',
             linestyle='--',
             label='Tendência')

    # Personalização
    plt.title(f'{nome_loja} - Relação: Vendas vs Faturamento por Categoria', fontsize=16, pad=20)
    plt.xlabel('Quantidade de Vendas', fontsize=12)
    plt.ylabel('Faturamento Total (R$)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.3)

    # Barra de cores (legend)
    cbar = plt.colorbar(scatter)
    cbar.set_label('Faturamento (R$)', fontsize=10)

    plt.legend()
    plt.tight_layout()
    plt.savefig(f'dispersao_vendas_faturamento_{nome_loja}.png', dpi=300, bbox_inches='tight')
    plt.show()

# Exemplo de uso:
grafico_dispersao_vendas_faturamento(loja, 'Loja 1')
grafico_dispersao_vendas_faturamento(loja2, 'Loja 2')
grafico_dispersao_vendas_faturamento(loja3, 'Loja 3')
grafico_dispersao_vendas_faturamento(loja4, 'Loja 4')
```

# 3. Média de Avaliação das Lojas

```python
def avaliacao_media(loja, nome_loja):

  media_loja = loja['Avaliação da compra'].mean()
  print(f'Média da {nome_loja}: {media_loja: .2f}')

avaliacao_media(loja, 'Loja 1')
avaliacao_media(loja2, 'Loja 2')
avaliacao_media(loja3, 'Loja 3')
avaliacao_media(loja4, 'Loja 4')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def gerar_grafico_avaliacoes(lojas_dados):
    """
    Gera um gráfico de barras com as médias de avaliação das lojas.

    Parâmetros:
    lojas_dados (list): Lista de tuplas com (DataFrame da loja, nome da loja)
    """
    nomes = []
    medias = []

    # Calcula as médias para cada loja
    for loja, nome in lojas_dados:
        media = loja['Avaliação da compra'].mean()
        nomes.append(nome)
        medias.append(media)
        print(f'Média da {nome}: {media:.2f}')  # Mostra no console também

    # Configuração do gráfico
    plt.figure(figsize=(10, 6))
    barras = plt.bar(nomes, medias, color=['#4CAF50', '#2196F3', '#FFC107', '#FF5722'])

    # Adiciona os valores nas barras
    for barra in barras:
        altura = barra.get_height()
        plt.text(barra.get_x() + barra.get_width()/2., altura,
                 f'{altura:.2f}',
                 ha='center', va='bottom', fontsize=12)

    # Personalização
    plt.title('Média de Avaliações por Loja', fontsize=16, pad=20)
    plt.xlabel('Lojas', fontsize=12)
    plt.ylabel('Média de Avaliação (0-5)', fontsize=12)
    plt.ylim(0, 5.5)  # Limite do eixo Y (avaliação vai até 5)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig('medias_avaliacoes.png', dpi=300)
    plt.show()

# Uso:
# (Substitua loja1, loja2... pelos seus DataFrames reais)
lojas = [
    (loja, 'Loja 1'),
    (loja2, 'Loja 2'),
    (loja3, 'Loja 3'),
    (loja4, 'Loja 4')
]

gerar_grafico_avaliacoes(lojas)
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_menos_vendidos(loja, nome_loja):
  """
  Função para verificar os produtos mais e menos vendidos em cada loja .

  Parâmetros:
  loja (DataFrame): O DataFrame contendo os dados da loja.
  nome_loja (str): Nome da loja para exibição.

  """
  vendas_produto = loja.groupby('Produto')['Preço'].sum()

  produto_mais_vendido = vendas_produto.sort_values(ascending=False).head(1)
  produto_menos_vendido = vendas_produto.sort_values(ascending=False).tail(1)

  print(f'\n{nome_loja}')
  print(f'Produto mais vendido: \n{produto_mais_vendido}')
  print(f'Produto menos vendido: \n{produto_menos_vendido}')

lojas = [(loja, 'Loja 1'),(loja2, 'Loja 2'), (loja3, 'Loja 3'), (loja4, 'Loja 4') ]

for loja, nome in lojas:
  produtos_mais_menos_vendidos(loja, nome)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analisar_produtos(loja, nome_loja):
    """
    Analisa produtos mais/menos vendidos por quantidade e faturamento.
    Retorna resultados e dados para gráficos.
    """
    # Análise por QUANTIDADE de vendas
    qtd_vendas = loja['Produto'].value_counts()
    mais_vendido_qtd = qtd_vendas.idxmax()
    menos_vendido_qtd = qtd_vendas.idxmin()

    # Análise por FATURAMENTO total
    faturamento = loja.groupby('Produto')['Preço'].sum()
    mais_vendido_fat = faturamento.idxmax()
    menos_vendido_fat = faturamento.idxmin()

    # Exibe resultados
    print(f'\n🔍 {nome_loja}:')
    print(f'▶ Mais vendido (quantidade): {mais_vendido_qtd} ({qtd_vendas[mais_vendido_qtd]} unidades)')
    print(f'▶ Menos vendido (quantidade): {menos_vendido_qtd} ({qtd_vendas[menos_vendido_qtd]} unidades)')
    print(f'▶ Mais vendido (faturamento): {mais_vendido_fat} (R${faturamento[mais_vendido_fat]:,.2f})')
    print(f'▶ Menos vendido (faturamento): {menos_vendido_fat} (R${faturamento[menos_vendido_fat]:,.2f})')

    return {
        'Loja': nome_loja,
        'Mais_Qtd': (mais_vendido_qtd, qtd_vendas[mais_vendido_qtd]),
        'Menos_Qtd': (menos_vendido_qtd, qtd_vendas[menos_vendido_qtd]),
        'Mais_Fat': (mais_vendido_fat, faturamento[mais_vendido_fat]),
        'Menos_Fat': (menos_vendido_fat, faturamento[menos_vendido_fat])
           }

lojas = [(loja, 'Loja 1'), (loja2, 'Loja 2'), (loja3, 'Loja 3'), (loja4, 'Loja 4')]

# Processa todas as lojas
resultados = [analisar_produtos(loja, nome) for loja, nome in lojas]

# Cria gráfico comparativo
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

# Gráfico 1: Quantidade de vendas
for res in resultados:
    ax1.bar(f"{res['Loja']}\nMais", res['Mais_Qtd'][1], color='#2ecc71')
    ax1.bar(f"{res['Loja']}\nMenos", res['Menos_Qtd'][1], color='#e74c3c')
ax1.set_title('PRODUTOS MAIS/MENOS VENDIDOS (QUANTIDADE)', fontsize=14)
ax1.set_ylabel('Unidades Vendidas')

# Gráfico 2: Faturamento
for res in resultados:
    ax2.bar(f"{res['Loja']}\nMais", res['Mais_Fat'][1], color='#3498db')
    ax2.bar(f"{res['Loja']}\nMenos", res['Menos_Fat'][1], color='#f39c12')
ax2.set_title('PRODUTOS MAIS/MENOS VENDIDOS (FATURAMENTO)', fontsize=14)
ax2.set_ylabel('Faturamento (R$)')

# Adiciona valores nas barras
for ax in (ax1, ax2):
    for bar in ax.patches:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:,.0f}' if ax == ax1 else f'R${height:,.2f}',
                ha='center', va='bottom')

plt.tight_layout()
plt.savefig('analise_produtos.png', dpi=300, bbox_inches='tight')
plt.show()

```

# 5. Frete Médio por Loja


```python
def media_loja_frete(loja, nome_loja):
    """
    Função para calcular e exibir a média de frete de uma loja.

    Parâmetros:
    loja (DataFrame): O DataFrame contendo os dados da loja.
    nome_loja (str): Nome da loja para exibição.
    """
    media_frete = loja['Frete'].mean()
    print(f'Média de Frete para {nome_loja}: {media_frete}')

lojas = [(loja, 'Loja 1'), (loja2, 'Loja 2'), (loja3, 'Loja 3'), (loja4, 'Loja 4')]

for loja, nome in lojas:
    media_loja_frete(loja, nome)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def calcular_medias_frete(lojas_dados):
    """
    Calcula médias de frete e gera gráficos comparativos

    Parâmetros:
    lojas_dados (list): Lista de tuplas (DataFrame, nome_loja)
    """
    # Cálculo das médias
    resultados = []
    for loja, nome in lojas_dados:
        media = loja['Frete'].mean()
        resultados.append({'Loja': nome, 'Média_Frete': media})
        print(f'Média de Frete - {nome}: R${media:.2f}')

    df_resultados = pd.DataFrame(resultados)


    # Gráfico de Barras
    plt.figure(figsize=(12, 6))
    barras = plt.bar(df_resultados['Loja'], df_resultados['Média_Frete'], color=cores)

    # Adiciona valores nas barras
    for bar in barras:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'R${height:.2f}',
                ha='center', va='bottom', fontsize=10)

    plt.title('Média de Valores de Frete por Loja', fontsize=14, pad=20)
    plt.xlabel('Lojas', fontsize=12)
    plt.ylabel('Média de Frete (R$)', fontsize=12)
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.savefig('media_frete_barras.png', dpi=300, bbox_inches='tight')
    plt.show()


calcular_medias_frete(lojas)
```
